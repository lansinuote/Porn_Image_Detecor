```python
import os
from PIL import Image

for file in os.listdir('pic'):
    if 'copy' in file or 'flip' in file:
        continue
    if file[0] == '0':
        #print('skip')
        continue
    print(file)
    im = Image.open('pic/' + file)
    im.save('pic/' + file[:-4] + '_copy.jpg')
    im = im.transpose(Image.FLIP_LEFT_RIGHT)
    im.save('pic/' + file[:-4] + '_flip.jpg')
    im.save('pic/' + file[:-4] + '_flip_copy.jpg')
```

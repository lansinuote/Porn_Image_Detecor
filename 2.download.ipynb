```python
import requests

with open('./wallhaven.csv') as f:
    lines = f.readlines()

for i in lines:
    i = i.strip().split(',')
    print(i)

    levels = i[0]
    types = i[1]
    url = i[2]

    content = requests.get(url=url).content
    
    fname = url.replace('https://th.wallhaven.cc/orig/', '').replace('/', '_')
    path = './pic/' + levels + '_' + types + '_' + fname
    

    with open(path, 'wb') as f:
        f.write(content)
```
